### Read in required data

```python
settings = load_settings()
NAME = settings['NAME']
df_tests = pd.read_pickle(f"data/{NAME}/statistical_tests/GRANK_ADJBMP.pkl")
df_tests = df_tests.drop(columns=["adj-GRANK p-value", "adj-GRANK statistic"])
```

### Set options

```python
# stop scientific notation
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

## Insiders are able to earn significant abnormal returns in the first weeks after disclosure.
This hypothesis is answered by taking a look at the statistics for purchases in table form

### Show in notebook

```python
results = df_tests.loc[:,"P - Purchase",:,:]

display(results)
```

### Print for Report

```python
print(results.style.to_latex(hrules = True
                             , position="H"
                             , position_float="centering"
                             , label="tab:purchase_statistics"
                             , caption="Statistics for Purchases over all investigation periods and CAR-periods."
                             , multirow_align="t"
                             , multicol_align="r"
                            )
     )
```
